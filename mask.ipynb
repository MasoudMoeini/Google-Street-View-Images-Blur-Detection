```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
img = cv2.imread('0-7.jpg')
cyan = np.full_like(img,(0,255,221))
mask= cv2.imread('7.jpg')
blend = 0.5
img_cyan = cv2.addWeighted(img, blend, cyan, 1-blend, 0)
result = np.where(mask==255, img_cyan, img)
plt.imshow(result,cmap='binary')
plt.axis("off")
plt.show()
```
